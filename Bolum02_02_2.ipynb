2) Cirq ile örnek

Cirq esasen devreler için kullanılır, ama Python’un doğal kompleks sayıları ve numpy ile görselleştirme kullanarak Argand düzlemi / polar gösterim kavramlarını gösterebiliriz.

```python
import numpy as np
import matplotlib.pyplot as plt
import cmath

# Örnek karmaşık sayı
z = 3 + 4j  # x = 3, y = 4

# Argand düzlemi için x ve y
x = z.real
y = z.imag

# Modül (norm)
r = abs(z)

# Argüman (faz açısı) – radians
theta = cmath.phase(z)

print(f"z = {z}")
print(f"Gerçel kısmı (x) = {x}")
print(f"Sanal kısmı (y) = {y}")
print(f"Modül r = |z| = {r}")
print(f"Argüman θ = arg(z) (radyan) = {theta}")
print(f"Polar form: z = r * (cos θ + i sin θ) = {r}*(cos({theta}) + i sin({theta}))")
print(f"Diğer polar gösterim: z = r * e^{iθ} = {r}*e^(i*{theta})")

# Görselleştirme – Argand düzlemi
plt.figure(figsize=(5,5))
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.scatter([x], [y], color='red')
plt.text(x, y, f"  z = {z}", fontsize=12)
plt.xlim(-r-1, r+1)
plt.ylim(-r-1, r+1)
plt.title("Argand Düzlemi")
plt.xlabel("Gerçel kısmı x")
plt.ylabel("Sanal kısmı y")
plt.grid(True)
plt.show()
```

```python
# Cirq ve gerekli kütüphaneleri yükleme
import cirq
import numpy as np
import matplotlib.pyplot as plt

# Örnek kompleks sayı
z = 3 + 4j

# Modül (r) ve Argüman (θ)
r = np.abs(z)
theta = np.angle(z)

print("Kompleks sayı (kartesyen):", z)
print("Modül |z|:", r)
print("Argüman θ (radyan):", theta)

# Euler formülü doğrulama: re^{iθ} = r(cosθ + i sinθ)
z_euler = r * np.exp(1j * theta)
print("Euler formülü ile elde edilen:", z_euler)

# Argand düzleminde görselleştirme
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.quiver(0, 0, z.real, z.imag, angles='xy', scale_units='xy', scale=1, color="red")
plt.text(z.real, z.imag, f"{z}", fontsize=12, color="blue")
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel("Re(z)")
plt.ylabel("Im(z)")
plt.title("Argand Düzlemi (Cirq - NumPy ile)")
plt.grid(True)
plt.show()
```

3) QuTiP ile örnek

QuTiP’te durum vektörlerini de kullanabiliriz; ama burada karmaşık sayıyı doğrudan tanımlayıp, modül ve argümanını hesaplayacağız. Ayrıca Argand düzlemi görselleştirmesi de yapılacak.

```python
import numpy as np
import matplotlib.pyplot as plt
from qutip import Qobj

# Karmaşık sayı
a = 5
b = -2
z = a + b * 1j  # z = 5 - 2i

# Gerçel ve sanal kısımlar
x = a
y = b

# Modül ve argüman
r = np.sqrt(x**2 + y**2)
theta = np.arctan2(y, x)  # arctan2 ile tüm quadrantalarda doğru açı

print(f"z = {z}")
print(f"Gerçel kısmı (x) = {x}, Sanal kısmı (y) = {y}")
print(f"Modül r = {r}")
print(f"Argüman θ = arg(z) (radyan) = {theta}")
print(f"Polar form: z = r (cos θ + i sin θ) = {r}*(cos({theta}) + i sin({theta}))")
print(f"Polar form (Euler): z = r e^(iθ) = {r} * exp(i*{theta})")

# Görselleştirme Argand düzlemi
plt.figure(figsize=(5,5))
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.scatter([x], [y], color='blue')
plt.text(x, y, f"  z = {z}", fontsize=12)
plt.xlim(-r-1, r+1)
plt.ylim(-r-1, r+1)
plt.title("Argand Düzlemi (QuTiP örneği)")
plt.xlabel("Gerçel kısmı x")
plt.ylabel("Sanal kısmı y")
plt.grid(True)
plt.show()
```

```python
import qutip as qt
import numpy as np
import matplotlib.pyplot as plt

# Kompleks sayı
z = 2 - 3j

# Modül ve Argüman
r = abs(z)
theta = np.angle(z)

print("Kompleks sayı:", z)
print("Modül |z|:", r)
print("Argüman θ:", theta)

# Euler formülü ile kontrol
z_euler = r * np.exp(1j * theta)
print("Euler formülü ile:", z_euler)

# QuTiP'te durum vektörü ile göstermek (normalleştirme)
psi = z * qt.basis(2, 0) + 1 * qt.basis(2, 1)
print("Durum vektörü:\n", psi)
```

4) Qiskit ile örnek

Qiskit, devre odaklıdır; ama yine Python kompleks sayıları ile modul ve argüman hesaplamasını ve görselleştirmeyi gösterebiliriz.

```python
import numpy as np
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit
from qiskit.quantum_info import Operator
import cmath

# Örnek karmaşık sayı
z = -1 + 1j

# Gerçel ve sanal kısımlar
x = z.real
y = z.imag

# Modül ve argüman
r = abs(z)
theta = cmath.phase(z)

print(f"z = {z}")
print(f"x = {x}, y = {y}")
print(f"Modül |z| = {r}")
print(f"Argüman θ = arg(z) (radyan) = {theta}")
print(f"Polar form: {r}*(cos({theta}) + i sin({theta}))")
print(f"Polar form (Euler): {r} * e^(i*{theta})")

# Görselleştirme
plt.figure(figsize=(5,5))
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.scatter([x], [y], color='green')
plt.text(x, y, f"  z = {z}", fontsize=12)
plt.xlim(-r-1, r+1)
plt.ylim(-r-1, r+1)
plt.title("Argand Düzlemi (Qiskit örneği)")
plt.xlabel("Gerçel kısmı x")
plt.ylabel("Sanal kısmı y")
plt.grid(True)
plt.show()
```
