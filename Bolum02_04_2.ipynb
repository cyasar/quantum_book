1. Cirq Örneği (Kompleks Üstel ve Trigonometrik Fonksiyonlar)

```python
import cirq
import numpy as np

# Kompleks sayı tanımla
z = 1 + 2j

# Üstel fonksiyon
exp_z = np.exp(z)

# Sin ve Cos
sin_z = np.sin(z)
cos_z = np.cos(z)

# Qubit oluştur
q = cirq.LineQubit(0)
circuit = cirq.Circuit(
    cirq.rx(z.real)(q),   # Gerçel kısım için rotasyon
    cirq.ry(z.imag)(q)    # İmajiner kısım için rotasyon
)

print("e^z =", exp_z)
print("sin(z) =", sin_z)
print("cos(z) =", cos_z)
print("\nCirq devresi:\n", circuit)
```

2. QuTiP Örneği (Üstel, Logaritma ve Güç Fonksiyonları)

```python
import numpy as np
from qutip import Qobj

# Kompleks sayı
z = 2 + 3j

# Qobj ile temsil
z_q = Qobj([[z]])

# Üstel
exp_z = z_q.expm()

# Logaritma (NumPy üzerinden)
log_z = np.log(z)

# Kompleks üs
w = 1 - 1j
z_pow_w = np.exp(w * np.log(z))

print("Üstel e^z =\n", exp_z)
print("Log(z) =", log_z)
print("z^w =", z_pow_w)
```

3. Qiskit Örneği (Euler Formülü ve Argand Düzlemi Görselleştirme)

```python
import numpy as np
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit

# Kompleks sayı
z = 3 + 4j
r = np.abs(z)
theta = np.angle(z)

# Euler formülü ile polar gösterim
polar = r * (np.cos(theta) + 1j*np.sin(theta))

# Quantum circuit (faz açısı ile rotasyon)
qc = QuantumCircuit(1)
qc.rz(theta, 0)

print("Kartezyen:", z)
print("Polar form:", polar)

# Argand düzleminde gösterim
plt.figure(figsize=(5,5))
plt.axhline(0, color='gray', lw=0.5)
plt.axvline(0, color='gray', lw=0.5)
plt.quiver(0, 0, z.real, z.imag, angles='xy', scale_units='xy', scale=1, color='r')
plt.text(z.real, z.imag, f"{z}", fontsize=12)
plt.title("Argand Düzleminde Z")
plt.xlabel("Gerçel")
plt.ylabel("Sanal")
plt.grid(True)
plt.axis("equal")
plt.show()

qc.draw("mpl")
```
